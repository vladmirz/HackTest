```
from itertools import permutations
from pprint import pprint

weather = ['晴れ', 'くもり', '雨', '雪', 'くもり', '大雨', '大雪', '暴風雨', '暴風雪']
conjunction = ['のち', '時々', '一時']

result = [f'{a} {c} {b}' for a, b in permutations(weather, 2) for c in conjunction]
pprint(result)
```
